#### Using Lasso Regression As The Model Algorithm

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Lasso

#this part of the code will split the dataset and use Lasso Regression Model
#essentially presents the Training & Test set scores, as well as the Number of Features used
tuberdataencoded = pd.get_dummies(tuberdata, drop_first=True)

y = tuberdataencoded['TB_Incidence_Rate']
X = tuberdataencoded.drop(columns=['TB_Incidence_Rate'])

#the train test is split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

#prints the output
lasso10 = Lasso(alpha=10, max_iter=100000).fit(X_train, y_train)
print("Training set score: {:.2f}".format(lasso10.score(X_train, y_train)))
print("Test set score: {:.2f}".format(lasso10.score(X_test, y_test)))
print("Number of features used:", np.sum(lasso10.coef_ != 0))
```

```python
#this part of the code does the same job previously
#a change in the alpha value to note any differences
lasso1 = Lasso(alpha=1, max_iter=100000).fit(X_train, y_train)
print("Training set score: {:.2f}".format(lasso1.score(X_train, y_train)))
print("Test set score: {:.2f}".format(lasso1.score(X_test, y_test)))
print("Number of features used:", np.sum(lasso1.coef_ != 0))
```

```python
#this part of the code does the same job previously
#a change in the alpha value to note any differences
lasso001 = Lasso(alpha=0.01, max_iter=100000).fit(X_train, y_train)
print("Training set score: {:.2f}".format(lasso001.score(X_train, y_train)))
print("Test set score: {:.2f}".format(lasso001.score(X_test, y_test)))
print("Number of features used:", np.sum(lasso001.coef_ != 0))
```

```python
#this part of the code plots the Lasso Regression Model used onto the Training and Test set scores
#demonstrates the coefficient to better understand the trend within the Tubercuolosis Dataset
plt.plot(lasso10.coef_, 's', label="Lasso alpha=10")
plt.plot(lasso1.coef_, '^', label="Lasso alpha=1")
plt.plot(lasso001.coef_, 'v', label="Lasso alpha=0.01")

xlims = plt.xlim()
plt.hlines(0, xlims[0], xlims[1], colors='black')
plt.xlim(xlims)
plt.ylim(-25, 25)
plt.xlabel("Coefficient index")
plt.ylabel("Coefficient magnitude")
plt.legend(ncol=2, loc=(0, 1.05))
plt.show()
```

#### Decision Tree

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor

#train and test using the tuberdata loaded from the Tubercuolosis_Dataset.csv
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

#created tree as variable to create the DecisionTreeRegressor
tree = DecisionTreeRegressor(random_state=0)
tree.fit(X_train, y_train)

#print the accuracy on training and test set score
print("Accuracy on training set: {:.3f}".format(tree.score(X_train, y_train)))
print("Accuracy on test set: {:.3f}".format(tree.score(X_test, y_test)))
```

```python
#the DecisionTreeRegressor now looks into the max_depth of 30
tree = DecisionTreeRegressor(max_depth=30, random_state=0)
tree.fit(X_train, y_train)

#print the accuracy on training and test set score
print("Accuracy on training set: {:.3f}".format(tree.score(X_train, y_train)))
print("Accuracy on test set: {:.3f}".format(tree.score(X_test, y_test)))
```

```python
from sklearn.tree import export_graphviz
import graphviz

#the DecisionTreeRegressor with random_state of 0 and max_depth of 4
tree = DecisionTreeRegressor(max_depth=4, random_state=0)
tree.fit(X_train, y_train)

#created an export_graphviz onto tree.dot that will allow the features to be displayed in a decision tree
export_graphviz(tree, out_file="tree.dot", feature_names=X.columns, 
                impurity=False, filled=True)

#opens the tree.dot decision tree created and display it onto the output
with open("tree.dot") as f:
    dot_graph = f.read()
display(graphviz.Source(dot_graph))
```

```python
#prints the feature importances to inspect their value
print("Feature importances:")
print(tree.feature_importances_)
```

```python
#definition for plot_feature_importances_tuberdata that will be used across the assignment
def plot_feature_importances_tuberdata(model, feature_names):
    features_name = [f for f in feature_names if not ('Country_' in f or 'Gender_' in f)]
    
    importances = model.feature_importances_

    update = [importances[i] for i in range(len(feature_names)) if feature_names[i] in features_name]
    
    n_features = len(features_name)
    plt.barh(np.arange(n_features), update, align='center')
    plt.yticks(np.arange(n_features), features_name)
    plt.xlabel("Feature importance")
    plt.ylabel("Feature")
    plt.ylim(-1, n_features)
    plt.show()

#called the definition to showcase the features importance onto tree (Decision Tree)
tree = DecisionTreeRegressor(max_depth=4, random_state=0)
tree.fit(X_train, y_train)
plot_feature_importances_tuberdata(tree, X.columns)
```

#### Random Forest

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor

#train and test using the tuberdata loaded from the Tubercuolosis_Dataset.csv
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

#created forest as variable to create the RandomForestRegressor
forest = RandomForestRegressor(n_estimators=10, random_state=0)
forest.fit(X_train, y_train)

#print the accuracy on training and test set score
print("Accuracy on training set: {:.3f}".format(forest.score(X_train, y_train)))
print("Accuracy on test set: {:.3f}".format(forest.score(X_test, y_test)))
```

```python
#called the definition to showcase the features importance onto forest (RandomForestRegressor)
plot_feature_importances_tuberdata(forest, X.columns)
```

#### Gradient Boosted Regression Trees (Gradient Boosting Machines)

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingRegressor

#train and test using the tuberdata loaded from the Tubercuolosis_Dataset.csv
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

#created gbrt as variable to create the GradientBoostingRegressor
gbrt = GradientBoostingRegressor(random_state=0)
gbrt.fit(X_train, y_train)

#print the accuracy on training and test set score
print("Accuracy on training set: {:.3f}".format(gbrt.score(X_train, y_train)))
print("Accuracy on test set: {:.3f}".format(gbrt.score(X_test, y_test)))
```

```python
#the GradientBoostingRegressor now looks into the max_depth of 3
gbrt = GradientBoostingRegressor(random_state=0, max_depth=3)
gbrt.fit(X_train, y_train)

#print the accuracy on training and test set score
print("Accuracy on training set: {:.3f}".format(gbrt.score(X_train, y_train)))
print("Accuracy on test set: {:.3f}".format(gbrt.score(X_test, y_test)))
```

```python
#the GradientBoostingRegressor now looks into the learning_rate of 0.1
gbrt = GradientBoostingRegressor(random_state=0, learning_rate=0.1)
gbrt.fit(X_train, y_train)

#print the accuracy on training and test set score
print("Accuracy on training set: {:.3f}".format(gbrt.score(X_train, y_train)))
print("Accuracy on test set: {:.3f}".format(gbrt.score(X_test, y_test)))
```

```python
#for the features importance, GradientBoostingRegressor will stay max_depth of 3
gbrt = GradientBoostingRegressor(random_state=0, max_depth=3)
gbrt.fit(X_train, y_train)

#called the definition to showcase the features importance onto forest (RandomForestRegressor)
plot_feature_importances_tuberdata(gbrt, X.columns)
```
